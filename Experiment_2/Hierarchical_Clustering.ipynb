#Clustering method - Hierarchical Clustering

Dataset - Iris.csv

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('/content/Iris.csv')
data = dataset.iloc[:, [1, 2, 3, 4]].values
```

```python
dataset.head
```

```python
#importing the hierarchy class of the scipy.cluster library as shc. 
#The hierarchy class has a dendrogram method which takes the value returned by 
#the linkage method of the same class. The linkage method takes the dataset and 
#the method to minimize distances as parameters.

import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

#Visualization

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
