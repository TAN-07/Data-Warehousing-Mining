```python
import numpy as np
import pandas as pd # for data processing
dataset = pd.read_csv('/content/online_retail_II.csv')
```

```python
dataset.head
```

#Preparing the dataset

```python
dataset = dataset.dropna()
dataset.head
```

```python
dataset['Description'].head
```

```python
#Grouping together the entries with same invoice code
dataset = dataset.groupby(['Invoice'])['StockCode'].apply(list)
dataset = dataset.tolist()[0:7000]        #keeping dataset limited to first 7000 rows
```

#Apriori Implementation

```python
!pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary,columns=te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df,min_support=0.01,use_colnames=True)
```

```python
frequent_itemsets
```

#Association Rule Mining

```python
from mlxtend.frequent_patterns import association_rules
res1 = association_rules(frequent_itemsets,metric='confidence',min_threshold = 0.8)
res1
```

```python
res2 = res1[['antecedents','consequents','support','confidence','lift']]
res2
```

```python
max(res2['confidence'])
```

```python
res3 = res2[res2['support']>=0.012]
res3 = res3[res3['confidence']>=1].reset_index()
res3[:5]
```
